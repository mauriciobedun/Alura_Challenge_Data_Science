#### 1.2 Visualizar a distribuição da variável target: Churn

```python
fig = px.histogram(df, x="target")
fig.show()
```

```python
def bota_titulo(titulo):

    fig.update_layout(
        title_text = titulo,
        title_xanchor = "center",
        title_x = 0.5,
        autosize=False,
        width=600,
        height=600,
    )
```

```python
fig = px.histogram(df, x='target', color = 'target')

bota_titulo('Contagem de Churn')

fig.update_layout(legend_title = 'Churn')

fig.update_xaxes(title_text='Churn',
                 type = 'category')
fig.update_yaxes(title_text='Nº Clientes')

## altera o label da contagem
fig.for_each_trace(lambda t: t.update(hovertemplate=t.hovertemplate.replace("count", "Nº Clientes")))

fig.show()
```

```python
fig = px.pie(df, names='target')

bota_titulo('Percentual de Churn')

fig.update_layout(legend_title = 'Churn')

fig.for_each_trace(lambda t: t.update(hovertemplate=t.hovertemplate.replace("Churn", "Churn")))

fig.show()
```

Observamos que há um desbalanceamento na variável resposta. Em algoritmos de classificação, este é um ponto importante a ser observado, pois pode influenciar negativamente o desempenho do nosso modelo.

##### 1.3.8 Label Encoding

```python
# função que transforma os valores categóricos de uma coluna em valores numéricos
def label_encoding(categorical_col):
  # altera o tipo de "object" para o tipo "category"
  encode = categorical_col.astype('category')

  # atribui a cada categoria um valor numérico diferente
  return encode.cat.codes
```

```python
# copiando a base de dados, para caso eu faça alguma coisa errada
df_copy = df.copy()
```

```python
# listando as colunas categóricas
df_copy.select_dtypes(include=object).columns
```

```python
# retirei a coluna "cliente_ID", pois não irá fazer sentido o encode dela
cat_columns = ['target', 'GENERO', 'CÔNJUGE', 'DEPENDENTES', 'ASS_TEL',
       'ASS_MULTI_TEL', 'ASS_INTERNET', 'ASS_AD_SEGURANÇA', 'BACKUP',
       'PROTEÇÃO_DISPOSITIVO', 'SUP_TÉCNICO', 'TV_CABO', 'STREAMING',
       'CONTRATO', 'FATURA_ONLINE', 'FORMA_PAGAMENTO']

# aplicando o label encoding
for col in cat_columns:
  # print(col)
  # print('Antes')
  # print(df_copy[col].unique())
  # print('Depois')
  df_copy[col] = label_encoding(df_copy[col])
  # print(df_copy[col].unique())
  # print('---')
```

##### 1.3.9 Correlação entre as variavéis 

```python
# calculando a correlação
correlation = df_copy.corr()
```

```python
# aumentando o tamanho da figura para facilitar a visualização
fig_dim = (20, 20)
fig, ax = plt.subplots(figsize=fig_dim)

# annot mostra o valor da correlação em cada célula;
# fmt formata os valores para um float com duas casas decimais;
# Para facilitar a interpretação da graduação das cores, vmin e vmax foram 
# deixados nos extremos da matriz de correlação ([1, -1])
sns.heatmap(correlation, annot=True, fmt='.2f', ax=ax, vmin=-1, vmax=1)

# salvando o heatmap em uma imagem
fig.savefig('heatmap.png')
```

Conclusões que podemos tirar com esse gráfico, apelidado por mim de teia de aranha. Grande parte das desistências do plano são pessoas que não assinaram as opções adicionais como, segurança, backup, proteção, suporte técnico, TV a cabo, streaming de filmes e o contrato é mensal.
